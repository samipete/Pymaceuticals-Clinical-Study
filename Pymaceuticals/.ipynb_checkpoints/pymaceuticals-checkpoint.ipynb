### Analysis

- This data indicates that the most effective drug regimen may be Capomulin 
- The drug Ramacine shows similar summary statistics to Capomulin and should therefore be further investigated for comparable effectiveness
- By looking at the bar charts for the number of mice tested for each drug regimen, we can see that Capomulin and Ramcine show the most timepoints which indicates that mice on these regimens may live longer.

- Furthermore, the linear regression model for Capomulin shows that mouse weight has a relatively strong positive correlation with tumor size (pearson's r=0.84). This suggests that increase in weight may be correlated with larger tumor size.

 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
# Read the data and create the path
mouse_data = pd.read_csv("data/Mouse_metadata.csv")
study_results = pd.read_csv("data/Study_results.csv")


# Combine the data into a single dataset; minor formatting
combined_data = pd.merge(study_results, mouse_data, how="left", on="Mouse ID")
combined_data['Tumor Volume (mm3)'] = combined_data['Tumor Volume (mm3)'].round(1)

# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
mouse_count = len(combined_data['Mouse ID'].unique())
mouse_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_mice = combined_data[combined_data.duplicated(['Mouse ID', 'Timepoint'])]

# variable for duplicate mouse ID
dup_mice_id = dup_mice['Mouse ID'].unique()

#show duplicate iD
display(dup_mice_id)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_mouse_all = pd.DataFrame(combined_data.loc[combined_data['Mouse ID']=='g989', :])
dup_mouse_all= dup_mouse_all.sort_values(['Timepoint'], ascending=True)
dup_mouse_all
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = combined_data[combined_data['Mouse ID'].isin(dup_mice_id)==False]
clean_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice = len(clean_data['Mouse ID'].unique())
clean_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of: 
# mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

# Assemble the resulting series into a single summary DataFrame.
drug_reg = clean_data.groupby(['Drug Regimen'])

sum_stats = pd.DataFrame()

sum_stats['Mean Tumor Volume'] = drug_reg['Tumor Volume (mm3)'].mean().round(2)
sum_stats['Median Tumor Volume'] = drug_reg['Tumor Volume (mm3)'].median().round(2)
sum_stats['Tumor Volume Variance'] = drug_reg['Tumor Volume (mm3)'].var().round(2)
sum_stats['Tumor Volume Std. Dev.'] = drug_reg['Tumor Volume (mm3)'].std().round(2)
sum_stats['Tumor Volume Std. Err.'] = drug_reg['Tumor Volume (mm3)'].sem().round(2)

sum_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line.

agg_reg_stats = drug_reg.agg({'Tumor Volume (mm3)':['mean', 'median', 'var', 'std', 'sem']}).round(2)

agg_reg_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (*greatest/max*) timepoint for each mouse (groupby mouse ID and timepoint)
max_tp = clean_data.groupby(['Mouse ID'])['Timepoint'].max()

#reset index
max_tp = max_tp.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
tumor_tp_merge = max_tp.merge(clean_data, on=['Mouse ID', 'Timepoint'], how= 'left')
#(show first 5 to ensure merge)
tumor_tp_merge.head()
```

```python
# Put treatments (Capomulin, Ramicane, Infubinol, and Ceftamin) into a list for a for loop (and later for plot labels)
four_treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumvol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. (start loop) 
for treament in four_treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumvol_final = tumor_tp_merge.loc[tumor_tp_merge['Drug Regimen']==treament, 'Tumor Volume (mm3)']
    # add subset 
    tumvol_data.append(tumvol_final)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumvol_final.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = tumvol_final[(tumvol_final < lower_bound) | (tumvol_final < lower_bound)]
    print(f"{treament}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fliers = dict(marker='o', markerfacecolor='red', markersize=10)
plt.boxplot(tumvol_data, flierprops=fliers, labels = four_treatments)
plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Final Tumor Volume (mm3) for each Mouse')

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
b128_df = clean_data.loc[(clean_data['Drug Regimen']=='Capomulin') & (clean_data['Mouse ID']=='b128')]
plt.plot(b128_df['Timepoint'], b128_df['Tumor Volume (mm3)'])

plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin Treatment for Mouse ID: b128')

plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
avgw_capreg = clean_data.loc[(clean_data['Drug Regimen']=='Capomulin')]
avgw_capreg = avgw_capreg.groupby('Mouse ID').mean()

plt.scatter(avgw_capreg['Weight (g)'], avgw_capreg['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Vol.(mm3)')
plt.title('Avg Tumor Vol. vs. Mouse Weight for Capomulin')

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

print(f"The correlation coefficient between Weight(g) and Tumor Volume (mm3) is {round(st.pearsonr(avgw_capreg['Weight (g)'],avgw_capreg['Tumor Volume (mm3)'])[0],2)}")

plt.scatter(avgw_capreg['Weight (g)'], avgw_capreg['Tumor Volume (mm3)'])

d_slope, d_int, d_r, d_p, d_std_err = st.linregress(avgw_capreg['Weight (g)'], avgw_capreg['Tumor Volume (mm3)'])
d_fit = d_slope * avgw_capreg['Weight (g)'] + d_int
plt.plot(avgw_capreg['Weight (g)'],d_fit, 'red')

plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Vol.(mm3')
plt.title('Average Tumor Vol. vs. Mouse Weight for Capomulin')

plt.show()
```
